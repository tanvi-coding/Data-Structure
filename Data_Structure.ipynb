### **Data Types and Structures Questions**

1.What are data structures, and why are they important?

  ->Data structures are methods of storing and organizing data so that it       can be used efficiently.
  They are important because they make programs faster, more efficient, and easier to manage.

2.Explain the difference between mutable and immutable data types with examples.

  ->Mutable: Can be changed after creation.
  Examples: List, Dictionary, Set

  Immutable: Cannot be changed after creation.
  Examples: Tuple, String, Integer

3.What are the main differences between lists and tuples in Python?

  ->Lists are mutable, tuples are immutable.


* Lists are slower, tuples are faster.


* Lists use [], tuples use ().

4.Describe how dictionaries store data.

  ->Dictionaries store data in key-value pairs. Internally, Python uses a hash  table, which makes searching very fast.

5.Why might you use a set instead of a list in Python?

  ->Use a set when:

  
* The order of items does not matter  
* You need faster searching
*  You want unique values (no duplicates)

6.What is a string in Python, and how is it different from a list?

  ->A string is a sequence of characters.
  Differences:

 * Strings are immutable, lists are mutable
 * Strings store only characters, lists can store any data type

7.How do tuples ensure data integrity in Python?

  ->Tuples are immutable, so once created, the data cannot be changed.
  This protects the data from accidental changes and ensures data integrity.

8.What is a hash table, and how does it relate to dictionaries in Python?

  ->A hash table is a data structure that stores data using a hash function.
  Python dictionaries use a hash table internally to provide fast key-based access.

9.Can lists contain different data types in Python?

   ->Yes. Python lists can store multiple data types.
   Example: [10, "hello", 3.5, True]

10.Explain why strings are immutable in Python.

  ->Strings are immutable because:

  * They save memory

  * They make programs faster

  * They are safe to use as dictionary keys

  * They prevent accidental changes

11.What advantages do dictionaries offer over lists for certain tasks

  * Faster data access

  * Access is through keys, not indexes

  * Data is more meaningful (key + value)

  * Searching is much quicker

12.How do sets handle duplicate values in Python?

  ->Sets automatically remove duplicate values.
    Example: {1, 2, 2, 3} becomes {1, 2, 3}

13.Describe a scenario where using a tuple would be preferable over a list.

   ->Use a tuple when data should not be changed, such as:

   * Coordinates (x, y)

   * Days or months of the year

   * Fixed settings or constants

14.How does the “in” keyword work differently for lists and dictionaries?

   * For lists: checks if a value exists

   * For dictionaries: checks only keys, not values

   Example:
   "name" in {"name": "Aditi"} → True
   "Aditi" in {"name": "Aditi"} → False

15.Can you modify the elements of a tuple? Explain why or why not.

  ->No, you cannot modify tuple elements because tuples are immutable.
    You cannot add, delete, or change any value inside a tuple.

16.What is a nested dictionary? Give an example.

   ->A nested dictionary is a dictionary that contains another dictionary inside it.

    Example:

    student = {
    "name": "Rohan",
    "marks": {"math": 92, "science": 88}
}

17.Describe the time complexity of accessing elements in a dictionary.

   ->Accessing elements in a dictionary takes O(1) time (constant time) because of the hash table.

18.In what situations are lists preferred over dictionaries?

   * Lists are preferred when:

   * Order matters

   * You need numeric indexing

   * Data is simple

   * Keys are not required

19.Why are dictionaries considered unordered, and how does that affect data retrieval?

   ->Dictionaries are considered unordered because you cannot access items by index.
   You must always use the key to retrieve data, not position.

20.Explain the difference between a list and a dictionary in terms of data retrieval.

   * List: Retrieve data using an index number

   * Dictionary: Retrieve data using a key
     Dictionaries offer faster and more meaningful access.
